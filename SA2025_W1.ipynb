### 1. What is name of car that has the highest horsepower?

### 2. How many cars have mpg ≥ 35?

```python
#your code here
num_cars = len(df[df['mpg'] >= 35])
print(num_cars)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
# Filter the DataFrame
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_df['origin'].mode()[0]

print("Most common origin for selected cars:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here

# Filter cars from Japan
japan_cars = df[df['origin'] == 'japan']

# Calculate mean acceleration (rounded)
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print("Mean acceleration of cars from Japan:", mean_acceleration)

```

### 5. Which year had the highest average mpg?

```python
#your code here
# Group by year and calculate mean mpg
avg_mpg_per_year = df.groupby('model_year')['mpg'].mean()

# Get the year with the highest average mpg
year_with_max_avg_mpg = avg_mpg_per_year.idxmax()
max_avg_mpg = round(avg_mpg_per_year.max(), 2)

print(f"Year with the highest average mpg: {year_with_max_avg_mpg} ({max_avg_mpg} mpg)")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# Step 1: Calculate the median mpg
median_mpg = df['mpg'].median()

# Step 2: Filter cars with mpg above the median
above_median_mpg = df[df['mpg'] > median_mpg]

# Step 3: Calculate horsepower-to-weight ratio
above_median_mpg['hp_weight_ratio'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Step 4: Find the max ratio
max_ratio = above_median_mpg['hp_weight_ratio'].max()

# Step 5: Find car(s) with this max ratio
best_cars = above_median_mpg[above_median_mpg['hp_weight_ratio'] == max_ratio]

# Step 6: Print result
print("Car(s) with best horsepower-to-weight ratio among those with above-median mpg:")
print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_weight_ratio']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here

import seaborn as sns

# Calculate average mpg per year and origin
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot with Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))

# Scatterplot
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),  # Adjust size range for better visibility
    alpha=0.7,
    palette='Set2'
)

# Titles and labels
plt.title('Car Weight vs Horsepower Colored by Origin\n(Size indicates MPG)', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.legend(title='Origin', loc='upper right')

plt.grid(True)
plt.show()
```
